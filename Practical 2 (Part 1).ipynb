## Correlation

```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
x = np.random.randint(0,50,1000)
```

```python
x
```

```python
y = x + np.random.normal(0,10,1000)
y
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
%matplotlib inline
```

```python
matplotlib.style.use('ggplot')
```

```python
plt.scatter(x,y)
plt.show()
```

# Both increase positive correlation
# if one increase other decrease negative correlation

```python
np.corrcoef(x,y)
```

# This correlation in strong positive 

```python
z = 100 - np.random.randint(0,10,1000)
```

```python
np.corrcoef(x,z)
```

```python
plt.scatter(x,z)
```

```python
import pandas as pd
```

```python
df= pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b'] = df['a'] +np.random.normal(0,10,1000)# positive correlated with 'a'
df['c'] =100- df['a'] +np.random.normal(0,10,1000)# negatively correlated with 'a'
df['d'] =np.random.randint(0,50,1000) # not correlated with 'a'
```

```python
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
```
